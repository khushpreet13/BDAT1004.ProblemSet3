```python
#Steps 1, 2
import pandas as pd
import numpy as np
data = pd.read_csv("wind1.csv")
data.head()
```

```python
#Step 3
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])
#re-arrange cols
data = data[['Date','RPT','VAL','ROS','KIL','SHA','BIR','DUB','CLA','MUL','CLO','BEL','MAL']]
data.head()
```

```python
#Step 4
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,
data.Date,data.Date - pd.offsets.DateOffset(years=100))
data.head()
```

```python
#Step 5
newData = data.set_index("Date")
newData.index.astype("datetime64[ns]")
```

```python
#Step 6
newData.isnull().values.ravel().sum()
```

```python
#Step 7
nData = newData.count()
print("There are total ", nData.sum(), " Non-missing values.")
```

```python
#Step 8
x = newData.mean()
x
```

```python
#Step 9
def stats(x):
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(stats)
loc_stats
```

```python
#Step 10
day_stats = newData.apply(stats,axis=1)
day_stats.head()
```

```python
#Step 11
jan_data = newData[newData.index.month == 1]
jan_data.mean()
```

```python
#Step 12
print( "Yearly:\n", newData.resample('A').mean())
```

```python
#Step 13
print("Monthly:\n", newData.resample('M').mean())
```

```python
#Step 14
print("Weekly:\n", newData.resample('W').mean())
```

```python
#Step 15
stats1 = newData.resample('W').mean().apply(lambda x: [x.count(),x.min(),x.max(),x.mean(),x.std()])
print(stats1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data_year = pd.DataFrame(data,columns=['Year'])
data_comp = pd.DataFrame(data,columns=['Marriages_per_1000','Divorces_per_1000']) 
fig, ax = plt.subplots(figsize=(9, 9))
for column in data_comp:
    ax.plot(data['Year'], data_comp[column])
ax.set_title('The Number of Marriages and Divorces per Capita in the U.S. between 1867 and 2014')
ax.legend(data_comp)
plt.xlabel("Year")
plt.ylabel("Per Capita")
plt.show()
```

# Question 7

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
x=pd.DataFrame(data.loc[(data['Year'] == 1900)])
y=pd.DataFrame(data.loc[(data['Year'] == 1950)])
z=pd.DataFrame(data.loc[(data['Year'] == 2000)])
final = pd.concat([x, y, z])
plt.rcParams["figure.figsize"] = (9, 9)
final.plot(x="Year", y=["Marriages_per_1000", "Divorces_per_1000"], kind="bar")
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per capita in the U.S. between 1900, 1950, and 2000')
plt.show()
```
